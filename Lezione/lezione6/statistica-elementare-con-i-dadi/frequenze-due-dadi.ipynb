```python
import random

# Numero di lanci
num_lanci = 6_000_000

# Inizializza una lista per registrare le frequenze delle somme
freqs = [0] * 11

# dodici valori 0
print(freqs)

for i in range(num_lanci):
    dado1 = random.randint(1, 6)
    dado2 = random.randint(1, 6)
    somma = dado1 + dado2

    # Incrementa la frequenza corrispondente alla somma ottenuta
    freqs[somma - 2] += 1

print(freqs, sum(freqs))

for index, freq in enumerate(freqs):
    print(f"{index:>3}: {freq:>10}")
```

```python
import matplotlib.pyplot as plt

# Calcola le percentuali
percentuali = [(freq / num_lanci) * 100 for freq in freqs]

# Crea una lista dei valori possibili della somma (da 2 a 12)
somma_valori = list(range(2, 13))

# Crea un grafico a barre orizzontali
plt.barh(somma_valori, percentuali)

# Aggiungi etichette per gli assi e il titolo
plt.ylabel('Somma dei due Dadi')
plt.xlabel('Percentuale')
plt.title(f'Percentuale della Somma dei Dadi ({num_lanci} lanci)')

# Aggiungi le percentuali accanto alle barre
for i in range(11):
    plt.text(percentuali[i], somma_valori[i], f'{percentuali[i]:.2f}%', va='center')

# Mostra il grafico
plt.show()
```
