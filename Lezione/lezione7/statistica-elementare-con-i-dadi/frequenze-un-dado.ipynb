```python
import random

freqs = [0] * 6
print(freqs)

for i in range(6_000_000):
    # face = random.randint(1, 6)
    face = random.randrange(1, 7)

    freqs[face - 1] = freqs[face - 1] + 1


print(freqs)

for index, freq in enumerate(freqs):
    print(f"{index}: {freq:>10}")

```

```python
import matplotlib.pyplot as plt

# Crea un elenco di valori del dado (da 1 a 6)
dado_valori = list(range(1, 7))

# Crea un grafico a barre
plt.bar(dado_valori, freqs)

# Aggiungi etichette per gli assi e il titolo
plt.xlabel('Valore del Dado')
plt.ylabel('Frequenza')
plt.title('Frequenza del Lancio del Dado (6,000,000 di lanci)')

# Aggiungi i valori sopra ogni barra
for i in range(6):
    plt.text(dado_valori[i], freqs[i], str(freqs[i]), ha='center', va='bottom')

# Mostra il grafico
plt.show()
# plt.savefig('grafico.png')
```
