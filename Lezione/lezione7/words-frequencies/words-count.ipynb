```python
# Contiamo le parole in una stringa di testo

text = "this is a simple text with several words \
this is anothertext with more words"

words_count = {} # ho cambiato leggermente il nome del libro, sarebbe il contatore di parole
words = text.split() # trovo le parole

# è come nel libro, ho solo invertito la if
for word in words:
    if word not in words_count:
        words_count[word] = 1
    else:
        words_count[word] += 1

print(words_count) # stampo il dizionario
```

```python
# semplifico qualche step

text = "this is a simple text with several words \
this is anothertext with more words"
words = text.split()

words_count = {}

# cambio il modo di creare il dizionario semplificando il ciclo (uso il metodo get)
for word in words:
    words_count[word] = words_count.get(word, 0) + 1

print(words_count) # stesso risultato
```

```python
# uso la classe Counter del modulo collections per semplificare ancora

text = "this is a simple text with several words \
this is anothertext with more words"
words = text.split()

# qui creo il dizionario sfruttanto un oggetto di tipo Counter
from collections import Counter
#words_count = Counter(words)
words_count = dict(Counter(words))

print(words_count) # stesso risultato
```

```python
# voglio ordinare in senso decrescente di frequenza il dizionario
# lo sappiamo fare, l'abbiamo visto l'ultima volta, ma ripetiamo

text = "this is a simple text with several words \
this is anothertext with more words"

words = text.split()

# qui creo il dizionario sfruttanto un oggetto di tipo Counter
from collections import Counter
#words_count = Counter(words)
words_count = dict(Counter(words))

print(words_count) # stesso risultato

# ordino in senso decrescente di frequenza
sorted_words_count = dict(sorted(words_count.items(), key=lambda item: item[1], reverse=True))
print(sorted_words_count) # nuovo dizionario in cui le frequenze più alte appaiono prima
```

```python
# voglio fare un grafico a barre
import matplotlib.pyplot as plt

text = "this is a simple text with several words \
this is anothertext with more words"

words = text.split()

# qui creo il dizionario sfruttanto un oggetto di tipo Counter
from collections import Counter
#words_count = Counter(text.split())
words_count = dict(Counter(words))

print(words_count) # stesso risultato

sorted_words_count = dict(sorted(words_count.items(), key=lambda item: item[1], reverse=True))
print(sorted_words_count) # nuovo dizionario in cui le frequenze più alte appaiono prima


words = list(sorted_words_count.keys())
frequencies = list(sorted_words_count.values())

# Creo il diagramma a barre
plt.figure(figsize=(10, 6))
plt.bar(words, frequencies)
plt.xlabel('Parole')
plt.ylabel('Frequenza')
plt.title('Diagramma a Barre delle Frequenze delle Parole')
# Ruota le etichette sull'asse x per una migliore leggibilità
plt.xticks(rotation=45)

plt.show()
plt.plot

# il grafico è bruttino
# e non mi dice niente
```
